```python
import evaluate
accuracy = evaluate.load('accuracy')
```

```python
accuracy.description
```

```python
accuracy.compute(references=[0,0,1,0,3], predictions=[0,1,1,0,3])
```

```python
for ref, preds in zip([9,1,3,5,9], [3,1,3,6,9]):
    accuracy.add(references=ref, predictions=preds)
    print(ref, ',', preds)

print(accuracy.compute())
```
